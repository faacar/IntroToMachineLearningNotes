```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Ã–rnek veri
data = sns.load_dataset("flights") 
data = data.pivot("month", "year", "passengers")

# Heatmap
sns.heatmap(data=data, annot=True, fmt="d", cmap="YlGnBu")

plt.show()
```

```python
data = sns.load_dataset("flights") 
# swarm plot
sns.swarmplot(x="year",y="passengers",data=data)
plt.show()
```

```python
data = sns.load_dataset("flights")

# Line plot
sns.lineplot(x="year", y="passengers",data=data)

plt.show()
```

```python
data = sns.load_dataset("iris")

sns.swarmplot(x="species", y="petal_length", data=data)

plt.show()
```

```python
data = sns.load_dataset("titanic")

# Bar plot
sns.countplot(data=data, x="sex")

plt.show()
```

```python

data = pd.DataFrame({
    'age_group': np.repeat(['18-30', '31-45', '46-60', '61+'], 100),
    'feedback_rating': np.random.randint(1, 6, 400)
})

# FacetGrid oluÅŸturma
g = sns.FacetGrid(data, col="age_group", margin_titles=True)

# Grafik tipi
g.map(sns.histplot, "feedback_rating")
```

```python
np.random.seed(0)
data = pd.DataFrame({'A': np.random.randn(100),
                     'B': np.random.choice(['X', 'Y'], size=100),
                     'C': np.random.randint(0, 10, size=100)})

g = sns.FacetGrid(data, col='B', col_wrap=2)
g.map(sns.histplot, 'A')
```

```python
np.random.seed(0)
data = pd.DataFrame({'A': np.random.randn(100),
                     'B': np.random.choice(['X', 'Y'], size=100),
                     'C': np.random.randint(0, 10, size=100)})

# FacetGrid oluÅŸturarak grafiÄŸi ayarlayalÄ±m
g = sns.FacetGrid(data, col='B', col_wrap=2)
g.map(sns.lineplot, 'C', 'A')
```

```python
sns.set_theme(style="white")

df = sns.load_dataset("penguins")

g = sns.JointGrid(data=df, x="body_mass_g", y="bill_depth_mm", space=0)
g.plot_joint(sns.kdeplot,
             fill=True, clip=((2200, 6800), (10, 25)),
             thresh=0, levels=100, cmap="rocket")
g.plot_marginals(sns.histplot, color="#03051A", alpha=1, bins=25)
```

```python
import seaborn as sns

sns.set_theme(style="darkgrid")

df = sns.load_dataset("tips")

g = sns.JointGrid(data=df, x="total_bill", y="tip", space=0, marginal_ticks=10)
g.plot_joint(sns.kdeplot,
             fill=True, clip=((0, 60), (0, 10)),
             thresh=0, levels=100, cmap="mako")
g.plot_marginals(sns.histplot, color="#242526", alpha=1, bins=20)
```
