```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from plotly.offline import iplot
import plotly.graph_objs as go

timesData = pd.read_csv("/kaggle/input/university-rankings-2011/University_Rankings_20220204.csv")
timesData.head
```

```python
dataframe = timesData[timesData.year == 2015]
```

```python
x = dataframe['world_rank']
y = dataframe['teaching']
z = dataframe['research']
size = dataframe['citations']


fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, mode='markers',
                                   marker=dict(size=size, color='rgb(255,0,0)', colorscale='Viridis'))])
iplot(fig)
# Layout

```

```python
fig.update_layout(title='2015 Üniversite Sıralaması', scene=dict(xaxis_title='Dünya Sıralaması',
                    yaxis_title='Öğretme Skoru', zaxis_title='Araştırma Skoru'))

iplot(fig)
```
